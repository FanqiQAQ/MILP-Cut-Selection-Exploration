```python
from pyscipopt import Model
```

```python
model = Model("Knapsack_MILP")
items = ['A', 'B', 'C', 'D']
values = {'A': 5, 'B': 8, 'C': 3, 'D': 9}
weights = {'A': 3, 'B': 4, 'C': 2, 'D': 5}
capacity = 10
```

```python
x = {}
for i in items:
    x[i] = model.addVar(name=f"item_{i}", vtype="B")
model.setObjective(sum(values[i] * x[i] for i in items), "maximize")
model.addCons(sum(weights[i] * x[i] for i in items) <= capacity, name="CapacityConstraint")
```

```python
y = model.addVar(name="filler", vtype="C", lb=0)
model.setObjective(sum(values[i] * x[i] for i in items) + 0.5 * y, "maximize")
model.addCons(sum(weights[i] * x[i] for i in items) + y <= capacity, name="MixedCapacity")
```

```python
model.optimize()
```
